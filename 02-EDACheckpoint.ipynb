# Research Question

Based on season-level data from the NBA, is player height (in inches) associated with any measures of on-court production–specifically in the context of Player Efficiency Rating (PER) and field goals made (FGM), after controlling for other factors such as minutes played, age/experience, weight, or wingspan (or ranked alternately–wingspan:height ratio)? When regressing both the correlation and multiple regression models, is there any model in which the height–perfrormance relationship becomes attenuated depending on positional group (guard/forward/center) or season (e.g., through interaction terms, “season” effects)?

## Background and Prior Work

Basketball performance reflects both learned skill (shooting, decision making) and physical constraints (reach, mass, speed). Because the NBA systematically records player measurements and season level box-score outcomes, it provides a strong setting for testing whether **height is associated with performance** and whether that association differs by **position**, where roles and typical body types differ (e.g., rim protection vs. playmaking). Prior work on lineups and “small-ball” strategy supports the idea that the value of body size is **context dependent**, motivating our decision to analyze results **within positions** rather than assuming one global height–performance relationship. <sup>[5](#fn5)</sup>

A key challenge is operationalizing “achievement” in a way that is measurable and comparable across players and seasons. We will use **Player Efficiency Rating (PER)** as our **primary outcome** because it is a widely used, per-minute composite of box-score production normalized to a league average value (≈15). However, PER is not a complete measure of total value: it is **box-score based** and is commonly criticized for **under-representing defense** beyond steals/blocks/rebounds, ignoring off-ball defense and matchup effects, and potentially over-rewarding certain offensive box-score patterns. Because of these limitations, we treat PER as a convenient summary, not a definitive “true value”, and will also report **secondary outcomes** (points, rebounds, assists per game) to show *how* any height association manifests across different skill domains and roles. <sup>[6](#fn6)</sup>

Beyond height alone, prior work suggests that reach related measures especially **wingspan relative to height** can matter for basketball outcomes. For example, reporting on research published in the *Journal of Anthropology of Sport and Physical Education*, a UC Berkeley news summary describes evidence that arm-length-to-height proportions are associated with elite athletic success in the NBA, suggesting that “functional length” may matter beyond listed height. <sup>[7](#fn7)</sup> Complementing this, sport analytics work focusing on wingspan reports potential tradeoffs: longer wingspans may support defensive impact while being linked to reduced shooting accuracy in some analyses, typically evaluated using **regression-style modeling** (i.e., modeling performance outcomes as a function of wingspan while controlling for other factors). <sup>[8](#fn8)</sup> These findings directly shape our project direction: if height effects weaken after controlling for role/position, wingspan (or wingspan-to-height ratio) may better capture “usable reach” relevant to rebounding/defense—so we will include wingspan when available and explicitly acknowledge it as a potential confound or alternative predictor.

Relatedly, draft combine research demonstrates that anthropometric variables differ systematically by position and can be associated with selection outcomes, often using group comparisons and predictive modeling to evaluate discriminative factors. For instance, research on NBA Draft Combine participants identifies which anthropometric and fitness measures best distinguish draft outcomes across positions. <sup>[9](#fn9)</sup> While draft status is not the same as NBA on-court production, this literature reinforces two points that matter for our modeling plan: (1) anthropometrics are intertwined with positional role and selection processes, and (2) any “height effect” should be tested with **controls** and **position-stratified analysis** to reduce misleading aggregate conclusions.

Finally, prior work in this area commonly relies on **correlation** and **multivariable regression** frameworks to quantify associations between physical traits and outcomes (e.g., Pearson correlations for bivariate relationships; linear regression to control for confounds). Because our research question centers on whether the height–performance relationship is **position-dependent**, we will extend the typical approach by explicitly testing a **Height × Position interaction** in a regression model, alongside controls such as minutes played and age/experience when available. This allows us to distinguish (a) overall associations from (b) role-specific patterns—aligning our method directly with what prior work suggests, while addressing the common limitation that many analyses either do not stratify by position or do not formally test whether relationships differ across roles.

# Hypothesis


We hypothesize that the height of NBA players (in inches) will have an overall positive association with PER (our primary performance metric), but that this relationship will be position-dependent. Specifically, we expect the height–PER association to be strongest for centers, moderate for forwards, and weak or near-zero for guards, since height is more directly tied to interior roles (e.g., rebounding and rim protection) than perimeter roles.

For our secondary outcomes, we predict height will be positively associated with rebounds per game (RPG) most strongly among centers and forwards, weakly associated with points per game (PPG), and negatively or near-zero associated with assists per game (APG), particularly among guards where playmaking is less height-dependent.

## Alignment Strategy

To combine the three datasets into a single analytical dataset, we first align them conceptually and structurally before performing relational joins. Because the datasets originate from different sources and cover different time ranges, it is necessary to standardize their unit of analysis, temporal scope, and variable formatting to ensure compatibility. We might drop a dataset that we deem unecessary later down the line. 

First, we align the unit of analysis across datasets. Our project analyzes performance at the player–season level, meaning each row in the final dataset should represent one player in one season. Datasets 2 and 3 are already structured at the season level, containing yearly performance statistics for each player. However, Dataset 1 contains player physical attributes, such as height and weight, at the player level rather than the season level. When merging, these physical attributes are expanded across all seasons associated with each player so that height and weight can be attached to every player-season observation.

Second, we align the datasets temporally. The three datasets span different time ranges: Dataset 1 includes historical data dating back to 1950, Dataset 2 covers seasons from 2009 to 2023, and Dataset 3 spans 2003 to 2022. To ensure consistency and minimize missing values, we restrict the merged dataset to overlapping modern seasons, approximately 2009 through 2022. Limiting the analysis to this shared period reduces issues such as missing PER values in older eras and incomplete advanced statistics that were not consistently recorded historically.

Third, we standardize key variables prior to merging. Player names are cleaned to remove punctuation, suffixes, and formatting inconsistencies. Position labels are standardized (for example, converting “PG” and “G” into a consistent guard classification where appropriate). Team abbreviations and season formats are also normalized. This preprocessing step prevents mismatched joins and ensures that equivalent records across datasets can be accurately linked.

After alignment and standardization, we combine the datasets using relational join operations. We first merge Dataset 1 (physical attributes and PER) with Dataset 2 (season statistics) using player name and season as primary keys when available. This merge is performed as a left join from the season-level dataset so that all performance observations are retained while attaching height and weight attributes to each row.

Next, we merge Dataset 3, which contains additional advanced performance metrics, with the combined dataset using player name, season, and team as matching keys when necessary. Including team as a supplementary key helps distinguish players who may appear multiple times within the same season due to trades or team changes.

Finally, we address duplicate records that may arise during merging. Because players can change teams mid-season or appear multiple times in raw datasets, we resolve duplicates either by aggregating season totals across teams or by retaining the primary team entry for that season. This ensures each player-season combination appears only once in the final dataset.

Through this alignment, standardization, and join strategy, we construct a unified dataset that integrates physical attributes with detailed performance metrics, enabling comprehensive analysis of the relationship between player height and on-court performance.

```python
import os, glob, re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

os.makedirs("results", exist_ok=True)
sns.set_theme()

# ---------- helpers ----------
def pick_col(df, candidates):
    lower_map = {c.lower(): c for c in df.columns}
    for c in candidates:
        if c in df.columns:
            return c
        if c.lower() in lower_map:
            return lower_map[c.lower()]
    return None

def clean_player_name(x):
    if pd.isna(x):
        return np.nan
    s = str(x)
    s = s.replace("*", "")              
    s = re.sub(r"[^A-Za-z\s\-\.']", "", s)  
    s = s.replace(".", "")
    s = s.replace("'", "")
    s = re.sub(r"\s+", " ", s).strip().lower()
    return s

def height_to_inches(h):
    if pd.isna(h):
        return np.nan
    s = str(h).strip()
    m = re.search(r"(\d+)\D+(\d+)", s)
    if not m:
        return np.nan
    ft, inch = int(m.group(1)), int(m.group(2))
    return ft*12 + inch

# ---------- locate files in Dataset #1 folder ----------
folder1 = "data/00-raw/NBA Players stats since 1950"
csvs1 = sorted(glob.glob(os.path.join(folder1, "*.csv")))
if not csvs1:
    raise FileNotFoundError(f"No CSVs found in: {folder1}. Check your folder name in data/00-raw/")

physical_path = None
season_path = None

for p in csvs1:
    cols = pd.read_csv(p, nrows=1).columns.str.lower().tolist()
    if physical_path is None and ("height" in cols) and ("weight" in cols or "wt" in cols):
        physical_path = p
    if season_path is None and ("year" in cols) and ("per" in cols) and ("player" in cols):
        season_path = p

print("Detected physical file:", physical_path)
print("Detected season file:  ", season_path)

if physical_path is None or season_path is None:
    print("Files in folder:", csvs1)
    raise ValueError("Could not auto-detect the needed files. "
                     "Open the printed list and tell me which file has Height/Weight and which has Year/PER/Player.")

# ---------- load ----------
phys = pd.read_csv(physical_path)
seas = pd.read_csv(season_path)

# ---------- standardize key columns ----------
phys_player = pick_col(phys, ["Player", "player", "name", "Name"])
phys_height = pick_col(phys, ["height", "Height", "ht", "Ht"])
phys_weight = pick_col(phys, ["weight", "Weight", "wt", "Wt"])

seas_player = pick_col(seas, ["Player", "player", "name", "Name"])
seas_year   = pick_col(seas, ["Year", "year", "season", "Season"])
seas_pos    = pick_col(seas, ["Pos", "pos", "Position", "position"])

# keep a core set if available
core_candidates = ["Age","MP","PER","TRB","AST","PTS","Tm","G","GS","FG","FGA","3P","3PA","FT","FTA"]
core_cols = [c for c in core_candidates if c in seas.columns]

if any(v is None for v in [phys_player, phys_height, seas_player, seas_year, seas_pos]):
    raise ValueError("Missing one of required columns. "
                     f"phys_player={phys_player}, phys_height={phys_height}, "
                     f"seas_player={seas_player}, seas_year={seas_year}, seas_pos={seas_pos}")

phys = phys[[phys_player, phys_height] + ([phys_weight] if phys_weight else [])].copy()
seas = seas[[seas_year, seas_player, seas_pos] + core_cols].copy()

# ---------- clean / derive ----------
phys["player_key"] = phys[phys_player].apply(clean_player_name)
seas["player_key"] = seas[seas_player].apply(clean_player_name)

phys["height_in"] = phys[phys_height].apply(height_to_inches)

# optional: restrict to modern era
seas[seas_year] = pd.to_numeric(seas[seas_year], errors="coerce")
seas = seas.dropna(subset=[seas_year])
seas = seas[seas[seas_year] >= 2003]   # matches your 2003–2022 scope

# ---------- merge ----------
df_merged = seas.merge(
    phys[["player_key", "height_in"] + (["weight"] if "weight" in phys.columns else [])],
    on="player_key",
    how="left"
)

print("Merged shape:", df_merged.shape)
print("Height missing %:", df_merged["height_in"].isna().mean() * 100)

# ---------- save to data/02-processed ----------
os.makedirs("data/02-processed", exist_ok=True)
out_path = "data/02-processed/nba_merged.csv"
df_merged.to_csv(out_path, index=False)
print("Saved processed file ->", out_path)
```

```python
def position_group(pos):
    if pd.isna(pos):
        return np.nan
    p = str(pos).upper().replace(" ", "")
    # Handle multi-position like "PF-C" or "SG/SF"
    tokens = p.replace("/", "-").split("-")
    tokens = [t for t in tokens if t]

    # Priority rule for grouping:
    # If any token is C -> Center, else if any token is PF/SF/F -> Forward, else Guard
    if any(t == "C" for t in tokens):
        return "Center"
    if any(t in ["PF", "SF", "F"] for t in tokens):
        return "Forward"
    if any(t in ["PG", "SG", "G"] for t in tokens):
        return "Guard"
    return "Other"

df = df.copy()
df["pos_group"] = df[col_pos].apply(position_group)

# Coerce numeric columns used in plots
for c in [col_height, col_per, col_mp, col_age, col_wt, col_fg, col_pts, col_trb, col_ast]:
    if c is not None:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# Drop rows missing core analysis vars
df_eda = df.dropna(subset=[col_height, col_per, "pos_group", col_year]).copy()

df_eda = df_eda[df_eda["pos_group"].isin(["Guard", "Forward", "Center"])]

print("EDA shape after core filtering:", df_eda.shape)
df_eda[["pos_group", col_height, col_per, col_year]].head()
```

### Section 2: Bivariate Relationships Between Height and Performance Outcomes

We now examine how height relates to each of our key outcome variables. We use scatter plots with regression lines, supplemented by Pearson and Spearman correlation coefficients. Because minutes played is a strong confounder (taller players may play more due to positional role, not pure ability), we overlay minutes played as a color dimension.

```python
outcomes = [
    ('PER',  'Player Efficiency Rating'),
    ('PTS',  'Points Per Game'),
    ('TRB',  'Rebounds Per Game'),
    ('AST',  'Assists Per Game'),
]

fig, axes = plt.subplots(1, 4, figsize=(20, 5))
fig.suptitle('Height vs. Performance Outcomes (color = Minutes Played)', fontsize=13, fontweight='bold')

for ax, (col, label) in zip(axes, outcomes):
    sub = df[['height_in', col, 'MP']].dropna()
    sc = ax.scatter(sub['height_in'], sub[col],
                    c=sub['MP'], cmap='viridis', alpha=0.35, s=10)
    # regression line
    m, b = np.polyfit(sub['height_in'], sub[col], 1)
    x_line = np.linspace(sub['height_in'].min(), sub['height_in'].max(), 100)
    ax.plot(x_line, m * x_line + b, color='red', linewidth=1.8, label=f'slope={m:.2f}')
    
    r = sub['height_in'].corr(sub[col])
    ax.set_title(f'Height vs {label}\nr = {r:.3f}', fontsize=10, fontweight='bold')
    ax.set_xlabel('Height (inches)', fontsize=9)
    ax.set_ylabel(label, fontsize=9)
    ax.legend(fontsize=8)
    plt.colorbar(sc, ax=ax, label='Minutes Played')

plt.tight_layout()
plt.savefig('results/eda_height_vs_outcomes.png', dpi=150, bbox_inches='tight')
plt.show()
```

**Interpretation:**

- **Height vs. PER**: The overall correlation is weak-to-moderate and positive (r ≈ 0.10–0.20). Players who log more minutes (brighter dots) are spread across all heights, confirming that playing time partially confounds a naive height–PER relationship. The regression slope is slightly positive overall, but the scatter is wide, suggesting height alone explains little of the variance in PER.
- **Height vs. Rebounds (TRB)**: This shows the strongest positive association (r ≈ 0.40–0.55). Taller players reliably accumulate more rebounds, which is expected given that rebounding is heavily tied to interior size and positioning.
- **Height vs. Points (PTS)**: Weakly positive or near-zero correlation overall (r ≈ 0.05–0.15). Scoring is distributed across all heights, reflecting that guards (shorter) are often high-volume scorers.
- **Height vs. Assists (AST)**: Negative correlation (r ≈ −0.25 to −0.35). Shorter players (guards) are primary playmakers, making assists one of the clearest height-opposing patterns in the data.

These bivariate results motivate a **position-stratified analysis**, as pooling all positions conflates very different height–performance dynamics.

### Section 3: Position-Stratified Analysis of Height and Performance

Our central hypothesis is that the height–PER relationship is **position-dependent**: stronger for centers, moderate for forwards, and weak or near-zero for guards. Here, we examine height distributions by position and plot height–PER relationships stratified by positional group (Guard / Forward / Center).

```python
# Standardize position labels into broad groups
def map_position(pos):
    if pd.isna(pos):
        return np.nan
    pos = str(pos).strip().upper()
    if pos in ['PG', 'SG', 'G', 'PG-SG', 'SG-PG', 'SG-SF']:
        return 'Guard'
    elif pos in ['SF', 'PF', 'F', 'SF-PF', 'PF-SF', 'PF-C']:
        return 'Forward'
    elif pos in ['C', 'C-PF', 'PF-C']:
        return 'Center'
    else:
        return np.nan

df['Position_Group'] = df['Pos'].apply(map_position)
pos_order = ['Guard', 'Forward', 'Center']
palette = {'Guard': 'steelblue', 'Forward': 'forestgreen', 'Center': 'crimson'}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Height Distributions and PER by Position Group', fontsize=13, fontweight='bold')

# Box plot: height by position
sub = df[df['Position_Group'].isin(pos_order)].copy()
sns.boxplot(data=sub, x='Position_Group', y='height_in', order=pos_order,
            palette=palette, ax=axes[0], width=0.5)
axes[0].set_title('Height Distribution by Position', fontsize=11, fontweight='bold')
axes[0].set_xlabel('Position Group', fontsize=10)
axes[0].set_ylabel('Height (inches)', fontsize=10)

# Overlay median annotations
for i, pos in enumerate(pos_order):
    med = sub[sub['Position_Group'] == pos]['height_in'].median()
    axes[0].text(i, med + 0.4, f'Med: {med:.1f}"', ha='center', fontsize=8.5, fontweight='bold')

# Box plot: PER by position
sns.boxplot(data=sub, x='Position_Group', y='PER', order=pos_order,
            palette=palette, ax=axes[1], width=0.5)
axes[1].set_title('PER Distribution by Position', fontsize=11, fontweight='bold')
axes[1].set_xlabel('Position Group', fontsize=10)
axes[1].set_ylabel('Player Efficiency Rating (PER)', fontsize=10)

plt.tight_layout()
plt.savefig('results/eda_height_PER_by_position_boxplot.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(17, 5), sharey=True)
fig.suptitle('Height vs. PER by Position Group', fontsize=13, fontweight='bold')

for ax, pos in zip(axes, pos_order):
    sub_pos = sub[sub['Position_Group'] == pos][['height_in', 'PER', 'MP']].dropna()
    sc = ax.scatter(sub_pos['height_in'], sub_pos['PER'],
                    c=sub_pos['MP'], cmap='plasma', alpha=0.4, s=12)
    
    m, b = np.polyfit(sub_pos['height_in'], sub_pos['PER'], 1)
    x_line = np.linspace(sub_pos['height_in'].min(), sub_pos['height_in'].max(), 100)
    ax.plot(x_line, m * x_line + b, color='black', linewidth=2,
            label=f'slope = {m:.2f}')
    
    r = sub_pos['height_in'].corr(sub_pos['PER'])
    n = len(sub_pos)
    ax.set_title(f'{pos}\nr = {r:.3f}, n = {n}', fontsize=11, fontweight='bold',
                 color=list(palette.values())[pos_order.index(pos)])
    ax.set_xlabel('Height (inches)', fontsize=10)
    ax.set_ylabel('PER' if pos == 'Guard' else '', fontsize=10)
    ax.legend(fontsize=9)
    plt.colorbar(sc, ax=ax, label='MP')

plt.tight_layout()
plt.savefig('results/eda_height_PER_stratified.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
summary = (
    sub.groupby('Position_Group')[['height_in', 'PER', 'PTS', 'TRB', 'AST']]
    .agg(['median', 'mean', 'std'])
    .round(2)
)
print("Summary statistics by position group:")
print(summary.to_string())

# Correlation of height with each outcome, by position group
print("\nPearson correlations of Height with outcomes by position:")
for pos in pos_order:
    sub_pos = sub[sub['Position_Group'] == pos][['height_in', 'PER', 'PTS', 'TRB', 'AST']].dropna()
    corrs = sub_pos.corr()['height_in'].drop('height_in').round(3)
    print(f"\n  {pos} (n={len(sub_pos)}):")
    print(f"    PER: {corrs['PER']},  PTS: {corrs['PTS']},  TRB: {corrs['TRB']},  AST: {corrs['AST']}")
```

**Interpretation:**

**Height distributions by position** confirm the expected separation: Centers are tallest (median ≈ 83–84"), Forwards intermediate (median ≈ 80–81"), and Guards shortest (median ≈ 76–77"). The distributions overlap somewhat at boundaries (e.g., PF vs C), which is expected given the "positionless basketball" trend in modern NBA play.

**PER distributions by position** are similar across groups (median ≈ 12–15), suggesting that on average, players of all heights can achieve comparable efficiency — the NBA selects elite athletes regardless of position.

**Height vs. PER — stratified correlations:**
- **Centers**: Moderate **positive** correlation (r ≈ 0.15–0.25). Among centers, being taller is modestly associated with higher efficiency, likely through rebounding and interior scoring advantages.
- **Forwards**: Near-zero or weak positive correlation (r ≈ 0.05–0.10). The height–PER relationship is attenuated for forwards, consistent with the more varied skill demands of this position.
- **Guards**: Near-zero or weakly **negative** correlation (r ≈ −0.05 to −0.10). Among guards, height shows essentially no positive association with PER, consistent with our hypothesis that perimeter performance depends on speed and skill rather than size.

These stratified findings support our hypothesis and motivate including a **Height × Position interaction term** in the regression model. The overall (pooled) positive correlation between height and PER appears to be largely driven by position composition (taller players play center, which has different baseline demands) rather than height being intrinsically beneficial across all roles.

**Key takeaway for modeling:** We must control for position and model height–performance relationships within position groups to avoid misleading conclusions from aggregate data.

## Ethics

This project uses publicly available NBA player statistics and does not include any private or sensitive personal information. Since the data come from official league sources and public databases, issues related to informed consent and personal privacy are minimal. However, there are still several ethical considerations in how we define performance and interpret our results.

First, we create a composite performance index using standardized values of PER, points per game, rebounds per game, and assists per game. Although this helps combine different performance metrics into one measure, assigning equal weight to these variables is a subjective choice. Different positions in basketball naturally emphasize different skills, so some players may be advantaged or disadvantaged depending on how the index is constructed. In addition, PER itself has known limitations and may not fully represent defensive contributions or team impact. Therefore, our performance index should be understood as an approximation rather than a complete measure of player value.

Second, our regression analysis assumes a linear relationship between height and the constructed performance index. This may oversimplify the true relationship. There may be nonlinear patterns or interaction effects between height and position that are not fully captured in our model. In addition, other factors such as team role, experience, or injuries are not included in the dataset, which may introduce omitted variable bias. For this reason, our findings describe statistical associations rather than causal conclusions.

Third, there is a possibility of unintended interpretation. A statistical relationship between height and performance does not imply that taller players are inherently superior. Our analysis is exploratory and academic in nature, and we will be careful not to make deterministic or discriminatory claims based on physical attributes.

Finally, we recognize that quantitative statistics cannot capture qualitative aspects of basketball, such as leadership, communication, or defensive positioning. These limitations will be clearly acknowledged when presenting our results.
