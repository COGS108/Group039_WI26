## Research Question

We seek to examine: Are NBA players' heights related to on-court performance, and does this relationship vary by position (guard, forward and center)? We’ll take height (inches) as the main predictor, and we want to measure its effect on achievement in terms of performance metrics (e.g., PER and FGM) (or related statistics; i.e. points, rebounds, assists). We will contrast how the height–achievement relationship applies across roles within position group to examine role specificity of the pattern. We will control for such potential confounds as weight, age/experience, minutes played, and wingspan and see if height still makes a difference. We’ll be using correlation and regression models, and visualizations (like scatterplots with trend lines) to illustrate it’s strength as well as direction of our relationship. The result will be an analytical answer to the question of whether taller players are, in fact, better and whether that is position-dependent.

## Background and Prior Work

Basketball performance depends on a combination of skill, tactics and physical characteristics, and theNBAis an ideal scenario as player measurements (e.g., height) and performance factors are recorded from several seasons. Because height affects what roles you play (rim protector, rebounder, shot contestor), and because it can vary systematically between positions in leagues, researchers and writers frequently discuss whether taller players really do achieve better outcomes—or if the benefits of being tall lie somewhere that’s mediated by position or how one plays, or even other things like arm length or conditioning. In this work, we examine the role of player height on performance/achievement in NBA players by season and position using basic statistical summaries of performance available for all NBA players. <sup>[1](#fn1)</sup>

One challenge is defining “achievement” in a measurable and comparable way. Most descriptions rely on a single-number summary that collapses many box-score contributions into an indication of overall impact. A common one that is referred to many times throughout the game now based on Cubs' broadcasts, is Player Efficiency Rating (PER) which takes good and bad stats and wraps them up into a per-minute stat designed to encapsulate efficiency. PER is one of the most widely used statistics in public basketball analytics because it can easily be compared between players and over seasons, but PER is nonetheless a model-based statistic with well-known limitations (e.g., it depends on a weighting scheme and box-score availability). The PER can be employed in our project as one endogenous variable, because it’s an effective cover to the total statistical contribution that is compared with height over various subgroups of players. <sup>[2](#fn2)</sup>

Previous work has also shown that anthropometric characteristics and body proportions can be associated with outcomes on the proxy of “success” in the NBA. For example, many studies of body proportions and NBA players report correlations between things like wingspan-to-height ratios and metrics such as draft position, while also making clear that loads of other factors (skill, role, team context) contribute to performance and career outcomes. <sup>[3](#fn3)</sup>

There are also similar works linking anthropometrics with strategy and lineup success. Zhang et al. explore how anthropometric factors can be used both to describe elite basketball lineups and performance patterns (including discussion around “small-ball” and lineup construction) underscoring the point that body size may matter differently based on a team’s deployment of players in relation to their role/s played. This supports our approach of considering results by position (i.e. guard/forward/center) instead of forcing one global relationship between height and achievement. <sup>[1](#fn1)</sup>

Finally, such questions are often examined in public analytics projects. Sivori’s study on height and free-throw percentage, for instance, presents a case where size is conditionally related to a skill outcome in some non-obvious way — and it serves as a useful textbook illustration of the kind of workflow you’d use to not-quite prove that sort of thing about NBA players. In contrast to that more specific skill-based outcome, our study examines broader types of performance (e.g., PER and potentially scoring-based metrics such as field goals made), and will also examine if the strength of relationships differs by position or season. <sup>[4](#fn4)</sup>
